```python
# --- Vorbereitung ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Darstellung verbessern
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# --- Daten einlesen ---
df = pd.read_csv("/home/o1d1/PycharmProjects/kaufland_wine_analysis/data/winequality-red.csv", sep=";")
df.head()

```

```python
# Überblick über die Spalten und Werte
df.info()    # Gibt Spaltennamen, Datentypen und fehlende Werte aus
df.describe() # Zeigt Durchschnitt, Minimum, Maximum, usw.
```

```python
#  Korrelationsmatrix und Balkendiagramm

# Korrelationen der originalen Spalten berechnen
corr_matrix = df.corr()

#Heatmap anzeigen
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='PuOr', fmt=".2f")
plt.title("Korrelationsmatrix der Merkmale")
plt.show()

# Korrelationen  berechnen und visualisieren
correlations = corr_matrix["quality"].drop("quality").sort_values(ascending=False)
correlations.plot(kind='bar', color='#106861')
plt.title("Korrelation der Merkmale mit der Weinqualität")
plt.ylabel("Korrelationskoeffizient")
plt.xlabel("Merkmal")
plt.tight_layout()
plt.show()
```

```python
#  Verteilung der Qualitätsstufen
sns.countplot(x='quality', data=df)
plt.title("Verteilung der Weinqualität")
plt.show()

# Grünton definieren
green_color = "#106861"

##  Boxplot: Alkoholgehalt nach Qualitätsstufe
sns.boxplot(x="quality", y="alcohol", data=df, color=green_color)
plt.title("Alkoholgehalt pro Qualitätsstufe")
plt.xlabel("Qualitätsbewertung")
plt.ylabel("Alkoholgehalt (%)")
plt.show()



```

```python
# Neue Zielvariable: gute Qualität = 1 (ab 7), sonst 0
df["good_quality"] = (df["quality"] >= 7).astype(int)

# Eingabedaten & Ziel
X = df.drop(["quality", "good_quality"], axis=1)
y = df["good_quality"]

# Train-Test-Split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Random Forest Modell
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Klassifikationsergebnisse
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Feature Importance
importances = model.feature_importances_
features = X.columns

# Richtung des Einflusses: Korrelation mit der Zielvariable 'quality'
correlations = df[features].corrwith(df["quality"])

# Farbe definieren je nach Vorzeichen der Korrelation
colors = ['#106861' if correlations[feat] >= 0 else '#D9534F' for feat in features]

# Nach Wichtigkeit sortieren
indices = np.argsort(importances)[::-1]
sorted_features = features[indices]
sorted_importances = importances[indices]
sorted_colors = [colors[i] for i in indices]

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x=sorted_importances,
    y=sorted_features,
    hue=sorted_features,            # <- notwendig für Farbanpassung
    palette=dict(zip(sorted_features, sorted_colors)),
    dodge=False,
    legend=False
)

plt.title("Wichtigkeit der Merkmale für Weinqualität")
plt.xlabel("Feature Importance")
plt.ylabel("Merkmal")
plt.tight_layout()
plt.show()
```
